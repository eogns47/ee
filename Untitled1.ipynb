```python
import random
print('구구단 게임 시작!\n')
r=0
w=0
a=1

while('true'):
    a=input('시작하려면 엔터,종료하려면 0 입력:')
    if(a=='0'):
        break
    num1=random.randrange(1,10)
    num2=random.randrange(1,10)
    print('{0:d}x{1:d}='.format(num1,num2),end='')
    answer=int(input())
    if(answer==num1*num2):
        r+=1
        print('정답!, 성공:{0:d}점, 실패:{1:d}점'.format(r,w))
    else:
        w+=1
        print('오답!, 성공:{0:d}점, 실패:{1:d}점'.format(r,w))
    

    
    

```

```python
import random
word=['cat','banana','frog','duck','fox','wolf','mouse']
input('타자게임 시작(엔터 입력):')
w=random.choice(word)
correct=0
wrong=0
total=0
while True:
    print('(종료 0):',w)
    my=input()
    if my=='0':
        break
    elif my==w:
        print('맞음!!')
        total+=1
        correct+=1
        w=random.choice(word)  
    else:
        print('오타! 다시 도전!')
        total+=1
        wrong+=1

print('총문제수:{0:d}, 맞은수:{1:d}, 정답률:{2:.1f}%'.format(total,correct,correct/total*100))
```
